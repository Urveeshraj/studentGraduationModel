```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define model names and metrics for each model without tuning
models_without_tuning = ['SVM', 'Random Forest', 'XGBoost', 'Bagging (Random Forest)', 'Bagging (SVM)', 'Extra Trees', 'AdaBoost', 'Gradient Boosting', 'Hybrid Voting', 'Stacking Ensemble']
accuracy_without_tuning = [92.73, 98.18, 94.55, 98.18, 94.55, 96.36, 96.36, 98.18, 98.18, 98.18]
precision_without_tuning = [96.15, 100.00, 100.00, 100.00, 96.30, 96.43, 96.43, 100.00, 100.00, 100.00]
recall_without_tuning = [89.29, 96.43, 89.29, 96.43, 92.86, 96.43, 96.43, 96.43, 96.43, 96.43]
f1_without_tuning = [92.59, 98.18, 94.34, 98.18, 94.55, 96.43, 96.43, 98.18, 98.18, 98.18]

# Define metrics for each model with tuning
models_with_tuning = ['Dynamic Bagging (SVM)', 'Dynamic Bagging (Random Forest)', 'AdaBoost', 'Gradient Boosting', 'Extra Trees', 'Hybrid Voting', 'Stacking Ensemble']
accuracy_with_tuning = [71.88, 75.00, 75.00, 78.12, 75.00, 78.12, 81.25]
precision_with_tuning = [83.33, 93.33, 93.33, 93.75, 84.21, 93.75, 94.12]
recall_with_tuning = [71.43, 66.67, 66.67, 71.43, 76.19, 71.43, 76.19]
f1_with_tuning = [76.92, 77.78, 77.78, 81.08, 80.00, 81.08, 84.21]

# Create a bar plot for each metric without tuning
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Model Performance Comparison Without Tuning', fontsize=16)

# Accuracy
sns.barplot(x=models_without_tuning, y=accuracy_without_tuning, ax=axes[0, 0], palette="Blues_d")
axes[0, 0].set_title("Accuracy (%)")
axes[0, 0].set_xticklabels(models_without_tuning, rotation=45, ha="right")

# Precision
sns.barplot(x=models_without_tuning, y=precision_without_tuning, ax=axes[0, 1], palette="Greens_d")
axes[0, 1].set_title("Precision (%)")
axes[0, 1].set_xticklabels(models_without_tuning, rotation=45, ha="right")

# Recall
sns.barplot(x=models_without_tuning, y=recall_without_tuning, ax=axes[1, 0], palette="Oranges_d")
axes[1, 0].set_title("Recall (%)")
axes[1, 0].set_xticklabels(models_without_tuning, rotation=45, ha="right")

# F1 Score
sns.barplot(x=models_without_tuning, y=f1_without_tuning, ax=axes[1, 1], palette="Purples_d")
axes[1, 1].set_title("F1 Score (%)")
axes[1, 1].set_xticklabels(models_without_tuning, rotation=45, ha="right")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Create a line plot for each metric with tuning
plt.figure(figsize=(12, 8))
plt.plot(models_with_tuning, accuracy_with_tuning, marker='o', label='Accuracy', color='blue')
plt.plot(models_with_tuning, precision_with_tuning, marker='o', label='Precision', color='green')
plt.plot(models_with_tuning, recall_with_tuning, marker='o', label='Recall', color='orange')
plt.plot(models_with_tuning, f1_with_tuning, marker='o', label='F1 Score', color='purple')

plt.title('Model Performance Comparison With Hyperparameter Tuning')
plt.xlabel('Models')
plt.ylabel('Performance Metrics (%)')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
