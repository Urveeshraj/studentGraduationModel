```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracy scores for each model
svm_accuracy = 0.9273 
rf_accuracy = 0.9818
xgb_accuracy = 0.9455

# Names of the models and corresponding accuracies
models = ['Support Vector Machine (SVM)', 'Random Forest', 'XGBoost']
accuracies = [svm_accuracy, rf_accuracy, xgb_accuracy]

# Plotting the accuracy comparison
plt.figure(figsize=(10, 6))
bar_width = 0.5
colors = ['skyblue', 'lightgreen', 'salmon']

bars = plt.bar(models, accuracies, color=colors, width=bar_width)
plt.ylim([0, 1])  # Ensure the y-axis goes from 0 to 1 for clarity
plt.xlabel('Machine Learning Models')
plt.ylabel('Accuracy Score')
plt.title('Model Accuracy Comparison')

# Adding text labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.02, f'{yval:.2%}', ha='center', va='bottom')

plt.show()
```
