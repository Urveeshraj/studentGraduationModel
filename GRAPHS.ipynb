```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for models without tuning
models_without_tuning = [
    "SVM", "Random Forest", "XGBoost", "Bagging (Random Forest)",
    "Bagging (SVM)", "Extra Trees", "AdaBoost", 
    "Gradient Boosting", "Hybrid Voting", "Stacking Ensemble"
]

accuracy_without_tuning = [92.73, 98.18, 94.55, 98.18, 94.55, 96.36, 96.36, 98.18, 98.18, 98.18]
precision_without_tuning = [96.15, 100.0, 100.0, 100.0, 96.30, 96.43, 96.43, 100.0, 100.0, 100.0]
recall_without_tuning = [89.29, 96.43, 89.29, 96.43, 92.86, 96.43, 96.43, 96.43, 96.43, 96.43]
f1_without_tuning = [92.59, 98.18, 94.34, 98.18, 94.55, 96.43, 96.43, 98.18, 98.18, 98.18]

# Data for models with tuning
models_with_tuning = [
    "Dynamic Bagging (SVM)", "Dynamic Bagging (Random Forest)", "AdaBoost",
    "Gradient Boosting", "Extra Trees", "Hybrid Voting", "Stacking Ensemble"
]

accuracy_with_tuning = [71.88, 75.00, 75.00, 78.12, 75.00, 78.12, 81.25]
precision_with_tuning = [83.33, 93.33, 93.33, 93.75, 84.21, 93.75, 94.12]
recall_with_tuning = [71.43, 66.67, 66.67, 71.43, 76.19, 71.43, 76.19]
f1_with_tuning = [76.92, 77.78, 77.78, 81.08, 80.00, 81.08, 84.21]

# Function to plot bar graphs for each metric
def plot_bar_graph(models, accuracy, precision, recall, f1, title):
    x = np.arange(len(models))
    width = 0.2  # Width of each bar
    
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Plotting each metric bar
    ax.bar(x - width*1.5, accuracy, width, label='Accuracy', color='skyblue')
    ax.bar(x - width/2, precision, width, label='Precision', color='salmon')
    ax.bar(x + width/2, recall, width, label='Recall', color='lightgreen')
    ax.bar(x + width*1.5, f1, width, label='F1 Score', color='gold')
    
    # Adding labels and title
    ax.set_xlabel('Models')
    ax.set_ylabel('Scores (%)')
    ax.set_title(f'{title} Performance Metrics')
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, ha="right")
    ax.legend()
    
    plt.tight_layout()
    plt.show()

# Function to plot line graph for each metric
def plot_line_graph(models, accuracy, precision, recall, f1, title):
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Plotting each metric line
    ax.plot(models, accuracy, marker='o', label='Accuracy', color='skyblue', linewidth=2)
    ax.plot(models, precision, marker='o', label='Precision', color='salmon', linewidth=2)
    ax.plot(models, recall, marker='o', label='Recall', color='lightgreen', linewidth=2)
    ax.plot(models, f1, marker='o', label='F1 Score', color='gold', linewidth=2)
    
    # Adding labels and title
    ax.set_xlabel('Models')
    ax.set_ylabel('Scores (%)')
    ax.set_title(f'{title} Performance Metrics')
    ax.legend()
    
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# Plot for models without tuning
plot_line_graph(models_without_tuning, accuracy_without_tuning, precision_without_tuning, recall_without_tuning, f1_without_tuning, "Without Tuning")

# Plot for models with tuning
plot_line_graph(models_with_tuning, accuracy_with_tuning, precision_with_tuning, recall_with_tuning, f1_with_tuning, "With Tuning")
```
